```python
import chromadb
import chromadb.utils.embedding_functions as embedding_functions

# Initialize Chroma client
chroma_client = chromadb.PersistentClient(path='.')

# Initialize Ollama embedding function
ollama_ef = embedding_functions.OllamaEmbeddingFunction(
    url="http://localhost:11434/api/embeddings",
    model_name="herald/dmeta-embedding-zh"
)

# Create or get collection in Chroma
collection = chroma_client.get_or_create_collection(
    name='activity',
    metadata={"hnsw:space": "cosine"},
    embedding_function=ollama_ef
)
```

```python
# Add documents to the collection
collection.add(
    documents=documents,
    metadatas=[{"data": "food"} for index in range(len(documents))],
    ids=[f"document-{index}" for index in range(len(documents))]
)
```

```python
# Perform a query on the collection
n_results = 2  # Define the number of results you want
# question = "請問我該怎麼把紅酒燉牛肉換成里肌肉？"
question = "請問我該怎麼做紅酒燉牛肉？"
results = collection.query(
    query_texts=[question],
    n_results=n_results,
    include=["documents", "metadatas", "distances"]
)

for content in results['documents'][0]:
    print(content)
```

```python
sentence = results['documents'][0]
prompt = f"""
你是一個食譜問答 AI 機器人，請使用台灣人習慣的繁體中文語句，透過知識直接回答問題。
回覆時，先感謝使用者的提問，然後回答答案。回答完問題最後告訴使用者，如果有任何問題，歡迎再次詢問。
標點符號請使用全形符號。\nKnowledge: {sentence}\nQuestion: {question}
"""

print("prompt", prompt)
print("___________________________")

response = ollama.chat(model='cwchang/llama-3-taiwan-8b-instruct', messages=[
  {
    'role': 'user',
    'content': prompt,
  },
])
print(response['message']['content'])
```
